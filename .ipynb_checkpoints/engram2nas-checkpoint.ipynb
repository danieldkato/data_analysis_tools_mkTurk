##### Define sessions to transfer, parameters:

```python
queried_sessions = [
    {'monkey' : 'Bourgeois', 'date' : '20251023'},
    {'monkey' : 'Bourgeois', 'date' : '20251022'},
    {'monkey' : 'Bourgeois', 'date' : '20251025'},
    {'monkey' : 'Bourgeois', 'date' : '20251026'}
]
queried_sessions = pd.DataFrame(queried_sessions)
```

##### Harcode base data path:

```python
base_data_path = os.path.join('/', 'mnt', 'smb', 'locker', 'issa-locker', 'Data') 
```

##### Check whether raw data directories for all sessions actually exist on Engram:

```python
sess_folders = find_sess_data_folders(base_data_path)
queried_sessions = pd.merge(queried_sessions, sess_folders, on=['monkey', 'date'], how='left')

# Raise warning about any sessions without raw data directory on Engram:
non_sess = queried_sessions[queried_sessions.path.isna()]
if non_sess.shape[0] > 0:
    S = non_sess.apply(lambda x : '{}, {}'.format(x.monkey, x.date), axis=1)
    warn_str = 'No raw data directory discovered for following sessions:\n' + '\n'.join(S)
    warnings.warn(warn_str)
    #print()

# Select only sessions with raw data directory on Engram:
queried_sessions = queried_sessions[~queried_sessions.path.isna()]
```

##### Get credentials:

```python
ip_address = input('IP Address:')
username = input('username:')
pw = input('PW:')
```

##### Establish SFTP connection:

```python
# Initialize client:
client = paramiko.SSHClient()
client.set_missing_host_key_policy(paramiko.AutoAddPolicy())
client.connect(ip_address, port=22, username=username, password=pw)

# Begin SFTP:
sftp = client.open_sftp()
```

##### Check whether any session directories already exist on remote server:

```python
# Exclude any monkeys with no directory on remote server:
remote_monkey_dirs = sftp.listdir()
monkeys = queried_sessions.monkey.unique()
dirs_per_monkey = np.array([len([d for d in remote_monkey_dirs if re.search(monkey, d) is not None]) for monkey in monkeys])
monkeys_missing_remote_dir = np.where(dirs_per_monkey == 0)[0] 
if len(monkeys_missing_remote_dir) > 1:
    warn_str = 'No remote directories found for following monkeys:\n' + '\n'.join(monkeys[monkeys_missing_remote_dir])
    warnings.warn(warn_str)
    queried_sessions = queried_sessions[~queried_sessions.monkey.isin(monkeys_missing_remote_dir)] 

#for s, session in sessions.iterrows():
```
